```python
import pandas as pd;

import numpy as np;

import geopandas as geo;

%matplotlib inline



 
```

```python


#housing unit value: https://data.wprdc.org/dataset/pittsburgh-american-community-survey-2015-miscellaneous-data/resource/13982f58-e618-400c-8f69-6a32dbd7c68f
# import dataset
Housing_value= pd.read_csv("housing_units.csv")
# filter to important info


print( Housing_value.groupby("Neighborhood"))



```

```python
Housing_value['Neighborhood']
```

```python

weight = Housing_value.groupby("Neighborhood").sum()['Id']
weight.sort_values(ascending=False)
#number of steps in lab 9 seems equivalent to 'id'

Housing_map = geo.read_file('Neighborhoods.shp')
Housing_map.plot()
```

```python
#how to make a weighted map: https://geopandas.org/en/stable/docs/user_guide/mapping.html


Housing_weight = Housing_map.merge(Housing_value, how = 'left', left_on= 'hood', right_on = 'Neighborhood')

#to copy and paste columns in a csv file you should print out the map once created and copy the identities. 
#Housing_weight  to call the csv text in full simply write the datatype without parameters
Housing_weight[['hood', 'Estimate; Total: - Less than $10,000', 'geometry']].head()

```

```python
Housing_weight.plot(column ='Estimate; Total: - Less than $10,000', legend = True)

#a surprising distribution of houses have units that are less than 10,000.
```

```python
Housing_2milormore = Housing_map.merge(Housing_value, how = 'left', left_on= 'hood', right_on = 'Neighborhood')

#Housing_2milormore #found the title  'Estimate; Total: - $2,000,000 or more' from calling 
Housing_weight[['hood', 'Estimate; Total: - $2,000,000 or more', 'geometry']].head()
```

```python
Housing_2milormore.plot(column = 'Estimate; Total: - $2,000,000 or more', legend = True)
#a roughly sparse but concentrated amount of regions have  houses that are 2 million or more. 
```
